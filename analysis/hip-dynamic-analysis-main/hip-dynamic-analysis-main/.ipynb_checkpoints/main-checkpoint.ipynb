```python
class Plotter:

    def __init__(self):
        #CANVA BG COLOR: '#fbf6f1'
        plt.style.use(['science', 'notebook', 'grid'])
        plt.rcParams['figure.figsize'] = [10, 6]
        plt.rcParams['axes.facecolor'] = '#FFFAF5'
        plt.rcParams['savefig.facecolor']='#fbf6f1'
        

        

    def plotOutput(self, calculation, patientComparison=None, winters=None, stressComparison=None):
        plt.clf()
        fg, ax = plt.subplots(1,1)
        fg
        ax.set_title(calculation.title_long)
        ax.set_xlabel("Gait Cycle [%]")
        ax.set_ylabel("{} [{}]".format(calculation.title, calculation.units))
        ax.grid(axis='y')

        ax.plot(calculation.x, calculation.output, '-', lw=2.5, ms=2, color='#348ABD')
        self.callout(calculation=calculation)
        if patientComparison:
           ax.plot(calculation.x, patientComparison.output, '-', lw=2.5, ms=2, color='#A60628')

        if winters:
            ax.legend(['Healthy Patient',"Winter's Data"], fontsize=10, fancybox=False, edgecolor='black')
        elif stressComparison:
            ax.legend(['Bone w/ Implant', 'Bone w/o Implant'],fontsize=10, fancybox=False, edgecolor='black')
        elif patientComparison:
            ax.legend(['Healthy Patient','Injured Patient'], fontsize=10, fancybox=False, edgecolor='black')
        else:
            ax.legend(['Healthy Patient'], fontsize=10, fancybox=False, edgecolor='black')

        ax.set_ylim(calculation.ylim)
        ax.set_xlim(0, 100)
        plt.savefig(f"graphs/{calculation.title_long}", dpi=300)
        plt.show()

    def plotThreeOutput(self, calculation1, calculation2, calculation3, patientComparison=None, winters=None, stressComparison=None):
        plt.clf()
        fg, ax = plt.subplots(1,3)
        fg
        ax.set_title(calculation.title_long)
        ax.set_xlabel("Gait Cycle [%]")
        ax.set_ylabel("{} [{}]".format(calculation.title, calculation.units))
        ax.grid(axis='y')

        ax.plot(calculation.x, calculation.output, '-', lw=2.5, ms=2, color='#348ABD')
        self.callout(calculation=calculation)
        if patientComparison:
           ax.plot(calculation.x, patientComparison.output, '-', lw=2.5, ms=2, color='#A60628')

        if winters:
            ax.legend(['Healthy Patient',"Winter's Data"], fontsize=10, fancybox=False, edgecolor='black')
        elif stressComparison:
            ax.legend(['Bone w/ Implant', 'Bone w/o Implant'],fontsize=10, fancybox=False, edgecolor='black')
        elif patientComparison:
            ax.legend(['Healthy Patient','Injured Patient'], fontsize=10, fancybox=False, edgecolor='black')
        else:
            ax.legend(['Healthy Patient'], fontsize=10, fancybox=False, edgecolor='black')

        ax.set_ylim(calculation.ylim)
        ax.set_xlim(0, 100)
        plt.savefig(f"graphs/{calculation.title_long}", dpi=300)
        plt.show()

    def callout(self, calculation, ax=None):
        ymin, ymax = min(calculation.output), max(calculation.output)
        xpos_min, xpos_max = np.argmin(calculation.output), np.argmax(calculation.output)
        x_min, x_max = calculation.x[xpos_min], calculation.x[xpos_max]
        
        text_min = "Frame {:.0f}, {:.3f} {units}".format(xpos_min+1, ymin, units=calculation.units)
        text_max = "Frame {:.0f}, {:.3f} {units}".format(xpos_max+1, ymax, units=calculation.units)
        
        if not ax:
            ax=plt.gca()

        bbox_props = dict(facecolor='#FFFAF5', edgecolor='black')

        arrowprops=dict(
                arrowstyle="-|>",
                connectionstyle="angle,angleA=0,angleB=90",
                color='k')

        kw = dict(
                xycoords='data',
                textcoords="axes fraction",
                arrowprops=arrowprops,
                bbox=bbox_props,
                ha="right",
                va="top")

        plt.annotate(text_max, xy=(x_max, ymax), xytext=(0.4,0.94), **kw)
        plt.annotate(text_min, xy=(x_min, ymin), xytext=(0.75,0.06), **kw)

plotter = Plotter()
plotter
```

```python
class Implant():

    def __init__(self):
        self.acetabulum = {
            'd': 53 #mm
        }
        self.cup = {
            'd': 54, #mm
            't': 54-43.68, #mm
            'v': 0.342,
            'E': 113800, #M/mm^2
            'material': 'Titanium Ti-6Al-4V'
        }

        self.liner = {
            'd': 43.58, #mm
            't': 43.58-36.05, #mm
            'E': 175000, #MPa
            'material': 'Zirconia',
            'v': 0.27,
        }

        self.head = {

            'd': 35.95,#mm
            'E': 175000, #N/mm^2
            'v': 0.27,
            'material': 'Zirconia'
        }

        self.neck = {
            'rLoad': np.array([20, 6]),
            't': 6, #mm
            'd': 12, #mm
            'yieldStrength': 970, #N/mm^2
            'E': 113800, #N/mm^2
            'material': 'Ti-6Al-4V'
        }

        self.stem = {
            'rLoad': np.array([66.704, -70.745]),
            't':15, #mm
            'dS':30, #mm
            'dM':15, #mm
            'b': 15,
            'h': 13.7,
            'yieldStrength': 970, #N/mm^2
            'material': 'PEEK',
            'dB': 30,
            'dC': 20,
            'Eb': 18.16, #GPa
            'Ec': 2.552, #GPa
            'Es': 4, #GPa
        }
        return

implant = Implant()
implant
```

```python
class Calculation:
    def __init__(self, patient, implant=None, title=None, title_long=None, units=None, ylim=None):
        self.title = title
        self.title_long = title_long
        self.units = units
        self.ylim = ylim
        self.output = np.empty(0)
        self.x = np.arange(0,106)
        self.x = 100*self.x/105
        self.patient = patient
        self.implant = implant
```

```python
class Patient:

    def __init__(self,winters_data,mass,height):

        #initialize patient properties
        self.mass = mass        #kg
        self.height = height    #m
        self.hipRx = []
        self.hipRy = []
        self.hipM = []

        #initialize dynamic analysis data to a dict
        self.dynamic = {}
        with open(winters_data, mode='r') as file:
            csv_reader = csv.DictReader(file)
            for row in csv_reader:
                frame = int(row['Frame'])
                del row['Frame']
                for key, value in row.items():
                    row[key] = float(value)
                self.dynamic[frame] = row

        #initialize Lengths, Masses, Weights
        self.L_foot = 0.152*self.height
        self.L_shank = (0.285-0.039)*self.height
        self.L_thigh = (0.53-0.285)*self.height
        self.m_foot = 0.0145*self.mass
        self.m_shank = 0.0465*self.mass
        self.m_thigh = 0.1*self.mass
        self.w_foot = self.m_foot*G
        self.w_shank = self.m_shank*G
        self.w_thigh = self.m_thigh*G
        
        for index in range(len(self.dynamic)):
            frame = index+1
            d = self.dynamic[frame] #d for data
            #geometry
            GRF_pos = [d["GroundCoFPX"],0]
            Ankle_pos = [d["AnkleX"]/100,d["AnkleY"]/100]
            Foot_CoM_pos = [d["FootCoMX"],d["FootCoMY"]]
            Knee_pos = [d["KneeX"]/100,d["KneeY"]/100]
            Leg_CoM_pos = [d["LegCoMX"],d["LegCoMY"]]
            Hip_pos = [d["HipX"]/100,d["HipY"]/100]
            Thigh_CoM_pos = [d["ThighCoMX"],d["ThighCoMY"]]
    
            #GRF = [100*d["GroundRX"]/56.7,100*d["GroundRY"]/56.7] # x100 /56.7 to normalize for the patient weight
            #GRF = np.array([(100/56.7)*d["GroundRX"], (100/56.7)*d["GroundRY"]])
            GRF = np.array([(mass/56.7)*d["GroundRX"],(mass/56.7)*d["GroundRY"]])
    
            #ankle frame
            F_ankle = [self.m_foot*d["FootAX"], self.m_foot*d["FootAY"]] - GRF - [0,self.w_foot]
            I_foot = ((0.475*self.L_foot)**2)*self.m_foot
            r_af = np.subtract(Ankle_pos,Foot_CoM_pos)
            r_gf = np.subtract(GRF_pos,Foot_CoM_pos)
            M_ankle = d["FootAA"]*I_foot - np.cross(r_af, F_ankle) - np.cross(r_gf, GRF)
            #WRONG
            
            #leg frame
            F_ankle = F_ankle*-1
            M_ankle = M_ankle*-1
    
            F_knee = [self.m_shank*d["LegAX"], self.m_shank*d["LegAY"]] - F_ankle - [0,self.w_shank] #good
            I_shank = ((0.302*self.L_shank)**2)*self.m_shank
            r_al = np.subtract(Ankle_pos,Leg_CoM_pos)
            r_kl = np.subtract(Knee_pos,Leg_CoM_pos)
            M_knee = d["LegAA"]*I_shank - np.cross(r_al, F_ankle) - np.cross(r_kl, F_knee) - M_ankle
            
            #thigh frame
            F_knee = F_knee*-1 #good
            M_knee = M_knee*-1 #good
    
            F_hip = [self.m_thigh*d["ThighAX"], self.m_thigh*d["ThighAY"]] - F_knee - [0, self.w_thigh]
            I_thigh = ((0.323*self.L_thigh)**2)*self.m_thigh #good
            #M_hip = d["ThighAA"]*I_thigh - np.cross(np.subtract(Knee_pos,Thigh_CoM_pos),F_knee) - np.cross(np.subtract(Hip_pos,Thigh_CoM_pos),F_hip) - M_knee
            M_hip = d["ThighAA"]*I_thigh - np.cross(np.subtract(Knee_pos,Thigh_CoM_pos),F_knee) - np.cross(np.subtract(Hip_pos,Thigh_CoM_pos),F_hip) - M_knee
            self.hipRx.append(F_hip[0])
            self.hipRy.append(F_hip[1])
            self.hipM.append(M_hip)

        self.hipRx = np.array(self.hipRx)
        self.hipRy = np.array(self.hipRy)
        self.hipM = np.array(self.hipM)
        self.hipResultant = (self.hipRx**2 + self.hipRy**2)**(1/2)
```

```python
hertzHeadLiner = Calculation(
        patient=healthyElder,
        implant=implant,
        title = "σ",
        title_long = "Dynamic Analysis: Hertz Contact Stress (Head-Liner)",
        units = "MPa",
        ylim = [5,42])

resultant = hertzHeadLiner.patient.hipResultant
c1 = hertzHeadLiner.implant.liner
c2 = hertzHeadLiner.implant.head
d1 = c1['d'] - c1['t']
d2 = c2['d']

Cg = (d1*d2)/(d1-d2)
Cm = (1-c1['v']**2)/c1['E'] + (1-c2['v']**2)/c2['E']
a = 0.721*(resultant*Cg*Cm)**(1/3)
Pmax = 1.5*resultant/(PI*(a)**2)

results.append(['Cg', Cg])
results.append(['Cm', Cm])
results.append(['a', a[FRAME]]) 
results.append(['Pmax', Pmax[FRAME]]) 

hertzHeadLiner.output = Pmax

plotter.plotOutput(calculation=hertzHeadLiner)
```

```python
#ignoring shear stresses, see bottom for why
neckStresses = Calculation(
        patient=healthyElder,
        implant=implant,
        title = "σ",
        title_long = "Maximum Compressive Stress on the Femoral Neck",
        units = "MPa",
        ylim = [-310,50])

resultant = neckStresses.patient.hipResultant
jointMoment = neckStresses.patient.hipM
neck = neckStresses.implant.neck

A = (1/4)*PI*neck['d']**2
I = (1/64)*(neck['d'])**4


#CALC IN REPORT USED +35, I USED -35 https://en.wikipedia.org/wiki/Rotation_matrix
#Added 8 degrees since thats what I measured as a normal standing abduction
theta = np.radians(-28)
#USE KNEE AND HIP POS TO DERIVE EXP FOR THETA_ABD
rotationMatrix = np.array([[np.cos(theta), -np.sin(theta)],
                           [np.sin(theta), np.cos(theta)]])
neckLoad = np.dot(rotationMatrix, np.vstack((-1*neckStresses.patient.hipRx, neckStresses.patient.hipRy)))

sigmaAxial, tau = neckLoad/A

results.append(['Neck sigma axial', sigmaAxial[FRAME]]) 

neckLoad = neckLoad.T
#factors in the moment generated by shear loading
M = 0-np.cross(neck['rLoad']/1000, neckLoad)
#not including the hip reaction momemt on Prof's advice
#otherwise, this matches up with the calcs in the document
sigmaBending = 1000*M*neck['t']/I

results.append(['Neck M', M[FRAME]]) 
results.append(['Neck sigma bending', sigmaBending[FRAME]]) 
neckStresses.output = sigmaAxial - abs(sigmaBending)
results.append(['Neck sigma MAX', neckStresses.output[FRAME]]) 
results.append(['Neck tau', tau[FRAME]]) 
SF = neck['yieldStrength']/abs(neckStresses.output.min())
print("The safety factor of the Femoral Neck is: {}.".format(SF))


plotter.plotOutput(calculation=neckStresses)
#quick maths, largest shear stress is only ~7MPa

```

```python
stemStresses = Calculation(
        patient=healthyElder,
        implant=implant,
        title = "σ",
        title_long = "Maximum Compressive Stress on the Femoral Stem",
        units = "MPa",
        ylim = [-115,20])


stem = stemStresses.implant.stem

theta = np.radians(-8)
#USE KNEE AND HIP POS TO DERIVE EXP FOR THETA_ABD
rotationMatrix = np.array([[np.cos(theta), -np.sin(theta)],
                           [np.sin(theta), np.cos(theta)]])
stemLoad = np.dot(rotationMatrix, np.vstack((-1*stemStresses.patient.hipRx, stemStresses.patient.hipRy)))

A = (PI/4)*(stem['dS']**2 - stem['dM']**2)
I = (1/64)*(stem['dS']**4 - stem['dM']**4)

[tau, sigmaAxial] = stemLoad/A
results.append(['Stem tau', tau[FRAME]]) 
results.append(['Stem axial', sigmaAxial[FRAME]]) 
stemLoad = np.vstack(stemLoad).T
M = 0 - np.cross(stem['rLoad']/1000, stemLoad)
results.append(['Stem M', M[FRAME]]) 
#again not including the hip reaction momemt on Prof's advice
#otherwise, this matches up with the calcs in the document

# x1000 to convert Nm to Nmm
sigmaBending = 1000*M*stem['t']/I
results.append(['Stem sigma bending', sigmaBending[FRAME]]) 
stemStresses.output = sigmaAxial - abs(sigmaBending)
results.append(['Stem sigma MAX', stemStresses.output[FRAME]]) 

plotter.plotOutput(calculation=stemStresses)
```

```python
#stress with implant
implantStresses = Calculation(
        patient=healthyElder,
        implant=implant,
        title = "Max Axial Stress",
        title_long = "Bone Stresses Compared",
        units = "MPa",
        ylim = [-110,25])

stem = implantStresses.implant.stem


#Geometry
Ib = (PI/64)*(stem['dB']**4 - stem['dC']**4)
Is = stem['b']*stem['h']**3/12
Ic = (PI/64)*stem['dC']**4 - Is

Ab = (PI/4)*(stem['dB']**2-stem['dC']**2)
As = stem['b']*stem['h']
Ac = (PI/4)*stem['dC']**2 - As


#Axial, Shear Stresses
theta = np.radians(-8)
#USE KNEE AND HIP POS TO DERIVE EXP FOR THETA_ABD
rotationMatrix = np.array([[np.cos(theta), -np.sin(theta)],
                           [np.sin(theta), np.cos(theta)]])
implantLoad = np.dot(rotationMatrix, np.vstack((-1*implantStresses.patient.hipRx, implantStresses.patient.hipRy)))


boneTau, sigmaBoneAxial = stem['Eb']*implantLoad / ( stem['Eb']*Ab + stem['Ec']*Ac + stem['Es']*As )
cementTau, sigmaCementAxial  = stem['Ec']*implantLoad / ( stem['Eb']*Ab + stem['Ec']*Ac + stem['Es']*As )
stemTau, sigmaStemAxial = stem['Es']*implantLoad / ( stem['Eb']*Ab + stem['Ec']*Ac + stem['Es']*As )

#Bending Stresses
implantLoad = np.vstack(implantLoad).T
M = 0 - np.cross(stem['rLoad']/1000, implantLoad)

results.append(['Bending Moment w/ Implant', M[FRAME]])

sigmaBoneBending = 1000*stem['Eb']*M*stem['t'] / ( stem['Eb']*Ib + stem['Ec']*Ic + stem['Es']*Is )
sigmaCementBending = 1000*stem['Ec']*M*stem['t'] / ( stem['Eb']*Ib + stem['Ec']*Ic + stem['Es']*Is )
sigmaStemBending = 1000*stem['Es']*M*stem['t'] / ( stem['Eb']*Ib + stem['Ec']*Ic + stem['Es']*Is )

results.append(['sigma bone axial w/ Implant', sigmaBoneAxial[FRAME]])
results.append(['sigma cement axial w/ Implant', sigmaCementAxial[FRAME]])
results.append(['sigma stem axial w/ Implant', sigmaStemAxial[FRAME]])

results.append(['sigma bone Bending w/ Implant', sigmaBoneBending[FRAME]])
results.append(['sigma cement Bending w/ Implant', sigmaCementBending[FRAME]])
results.append(['sigma stem Bending w/ Implant', sigmaStemBending[FRAME]])

results.append(['tau bone w/ Implant', boneTau[FRAME]])
results.append(['tau cement w/ Implant', cementTau[FRAME]])
results.append(['tau stem w/ Implant', stemTau[FRAME]])


#Combined, comparison with intact bone
sigmaBoneMax = sigmaBoneAxial - abs(sigmaBoneBending)
sigmaCementMax = sigmaCementAxial - abs(sigmaCementBending)
sigmaStemMax = sigmaStemAxial - abs(sigmaStemBending)

results.append(['sigma bone MAX w/ Implant', sigmaBoneMax[FRAME]])
results.append(['sigma cement MAX w/ Implant', sigmaCementMax[FRAME]])
results.append(['sigma stem MAX w/ Implant', sigmaStemMax[FRAME]])

implantStresses.output = sigmaBoneMax
plotter.plotOutput(calculation=implantStresses, patientComparison=stemStresses, stressComparison=True)
```

```python
#Stress Comparisons, Stress Shielding
implantFactor = Calculation(
        patient=healthyElder,
        implant=implant,
        title = "IF",
        title_long = "Implant Factor",
        units = "MPa / MPa",
        ylim = [0,3.5])

stressShielding = Calculation(
        patient=healthyElder,
        implant=implant,
        title = "SS",
        title_long = "Stress Shielding",
        units = "MPa / MPa",
        ylim = [0,1])

sigmaIntactBone = stemStresses.output
sigmaImplantBone = sigmaBoneMax

implantFactor.output = abs(sigmaIntactBone / sigmaImplantBone)
stressShielding.output = (sigmaIntactBone - sigmaImplantBone) / sigmaIntactBone

results.append(['Stress Comparison Factor', implantFactor.output[FRAME]])
results.append(['Stress Shielding Value', stressShielding.output[FRAME]])


print(stressShielding)

plotter.plotOutput(implantFactor)
plotter.plotOutput(stressShielding)

print(np.average(implantFactor.output))
print(np.average(stressShielding.output))
```
